1. У вільному доступі в телеграмі є курс по створенню робота для торговлі на binance. Даний робот включає у себе: 
    - всі механізми комунікації з біржею бінанс
    - стратегію визначення моменту входу в транзакцію купівлі або продажу на монеті ETH
    - механізми входу і виходу з транзакції
    - телеграм бот для керування роботом і відображення інформації роботи програми
    - розміщення програми на віддаленому сервері
    
    

2. До даної програми додав наступні механізми:
    - пошук точки входу на всіх представлених на біржі монетах (а не тільки на ETH)
    - логування всіх важливих механізмів (txt файл логів вже сягає 10 Мб)
    - трейлінг стоп (межа stop loss сліує за зміною ціни монети)
    
   Дуже важливо добавити csv файл, де представлена максимальна статистична інформація на момент входу 
   в транзакцію. Це необхідно для отримання матеріалу і виявлення залежностей між певними факторами і результатом
   даної транзакції. Дане дослідження і являється професійною задачою спеціаліста з data science.

3. В реалі програма виявилася не настільки прибутковою, тому до розміщення на сервері, працюю над різноманітними 
   модифікаціями стратегії і виявлення найбільш прибуткових. Крім того ці зміни одразу несуть за собою цілу купу багів, які    потрібно виявляти і фіксити. Зараз на мінімальних ставках, втрата на кожній неприбутковій угоді сягає декількох центів, 
   що дозволяє тестити робота без особливих затрат.

4. У будь якому випадку, це вже реальна робота програмістом з отриманням реального досвіду. Крім того програмування на        практиці з отриманням реальних результатів виявилося дуже захоплюючим зайняттям. 

5. Якщо когось ця тема зацікавить, то після запуску вашого власного робота і після отримання деякого досвіду, можна
   розвиватися далі, адже існують більш складні для математичного моделювання стратегії торговлі, на яких люди заробляють
   нормальні гроші.
